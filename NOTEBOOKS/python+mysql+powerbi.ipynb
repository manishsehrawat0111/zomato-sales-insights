```python
# drop unnamed: 20 column beacuse it is fully empty
df = df.drop(columns=['Unnamed: 20'])
```

```python
# drop cuisines column beacuse it is duplicate
df=df.drop(columns=["Cuisines"])
```

```python
# filled columns with unknown using fillna
df[['Cuisines 1','Cuisines 2','Cuisines 3',
  'Cuisines 4','Cuisines 5','Cuisines 6','Cuisines 7',
  'Cuisines 8']]=df[['Cuisines 1','Cuisines 2','Cuisines 3',
                     'Cuisines 4','Cuisines 5','Cuisines 6','Cuisines 7','Cuisines 8']].fillna("unknown")
```

```python
df.isnull().sum().sum()
```

```python
df["City"]=df["City"].astype('string')
print(df["City"].dtypes)
```

```python
df['Cuisines 1']=df['Cuisines 1'].astype('string')
print(df['Cuisines 1'].dtypes)
```

```python
df["Currency"]=df["Currency"].astype("string")
print(df["Currency"].dtypes)
```

```python
cuisines_coln=["Cuisines 2","Cuisines 3","Cuisines 4","Cuisines 5","Cuisines 6","Cuisines 7","Cuisines 8"]
df[cuisines_coln]=df[cuisines_coln].astype("string")
```

```python
df["Datekey_Opening"] = df["Datekey_Opening"].str.replace("_", "-", regex=False)
df["Datekey_Opening"] = pd.to_datetime(df["Datekey_Opening"], format="%Y-%m-%d")
```
